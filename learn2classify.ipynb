#  Neural Networks 101: Neurons and Activation Functions

A *neural network* is a network of neurons connected together (and *how they are connected* matters quite a lot!). 

Each neuron in a neural network has asscoiated with it a scalar-valued *activation function* $f_a(\cdot)$, along with a weight $w$ and a *bias* $b$.  

<img src=neuron.png width=400px alt="A neuron and its output">

The neuron represented in the figure above takes as its input a scalar $x$, and returns as the output

$$ f_1(x,\color{green}{w},\color{blue}{b}) = f_{a}(\color{green}{w}\,x + \color{blue}{b}).$$

Here $w$ is the weight of the input *edge* to the neuron.

## Linear networks

Many [activation functions](https://en.wikipedia.org/wiki/Activation_function) are used in neural networks; see [here](https://en.wikipedia.org/wiki/Activation_function#Comparison_of_activation_functions) for many examples.

$$\textrm{linear}(z)=z.$$

Linear neural networks use the linear activation function -- they are also powerful; their power comes from the choice of weights and biases, which are learned.

We can define this in Julia as in the next cell.

## Representing the output of a single neuron to a stream of inputs 

We would now like to mathematically represent the output of a single neuron to a stream of inputs.

To that end, we will use the  [dot operator](https://docs.julialang.org/en/v1/manual/functions/index.html#man-vectorized-1) in Juia Recall that if `X` is an `Array` then `f.(X)` is equivalent mathematically to applying the function `f` elementwise to every element of the array `X`.  

Consequently,  if $X$ is a $1 \times n $ array with 

$$X = \begin{bmatrix} X[1] & X[2] & \cdots &  X[n] \end{bmatrix},$$

then

$$f.(X) = \begin{bmatrix} f(X[1]) & f(X[2]) & \cdots &  f(X[n]) \end{bmatrix}.$$

This can be used to succinctly express the output of a neural network for a vector of inputs as described next. 

Indeed! In Julia, we can express the output of a neural to a vector of inputs as 

```julia
f_a.(w * x .+ b)
```

where the `.+ b` is equivalent to `+ b * ones(size(x))`, but avoids allocating memory for `ones(size(x))`.

# A simple neural network that learns to classify

<img src=multineuron1output.png width=450px alt="A multiple input neuron, single output neural network">

We will next examine a neural network having the above toplogy, with $n$ input neurons having weights $w_1, \ldots, w_n$ associated with the inputs $x_1, \ldots, x_n$. The output of this network is given by 

\begin{equation}
g_n(x,w,b) = f_{\rm a}\left(\sum_{i=1}^{n} w_i x_i +b\right),
\end{equation}

where $f_{\rm a}(z)$ is the activation function, $w_1, \ldots, w_n$ are the weights, and $b$ is the bias. Define the weight vector $w$ as 

$$ w = \begin{bmatrix} w_1 & \ldots w_n \end{bmatrix}^T,$$

and the vector of inputs $x$ as 

$$ x = \begin{bmatrix} x_1 & \ldots x_n \end{bmatrix}^T.$$

Then we can express the output as 

\begin{equation}
g_n(x,w,b) = f_{\rm a}( x^T w + b)~.
\end{equation}

Now that we've seen how to handle multidimensional inputs individually, let us consider how to handle many of them simultaneously. In many applications, we are given $N$ data samples of n-dimensional vectors or instances, and it is convenient to arrange them as columns of a  matrix:

$$ x = \begin{bmatrix} x[:,1] & \ldots & x[:,N] \end{bmatrix}.$$

In this notation, the $j$-th column of $x$ is $x[:,j]$, an $n$-dimensional vector corresponding to the $j$-th sample. 

We would like to compute $w^T x[:,j]$ for $j = 1, \ldots, N$. We can do so conveniently via the operation

$$ w^T x = \begin{bmatrix} w^T x[:,1] & \dots & w^T x[:,N] \end{bmatrix},$$

which produces a  $1 \times N$ matrix (or row vector) whose $j$-th element corresponds to the scalar $w^Tx[:,j]$ for $j = 1, \ldots, N$.

Note that

$$x^T w = \begin{bmatrix} x[:,1]^T w\\  \vdots \\ x[:,N]^T w \end{bmatrix},$$ 

which has the same elements as $w^Tx$ but *not* the same shape (rows or columns). This will matter, as we shall see shortly, because we will want to subtract matrices of the same shape.

In the next cell, define the function `gn` which takes as inputs
* the $n \times N$ array `x`,
* the vector `w` and scalar `b`, and
* the activation function `f_a`.

`gn` should return $g_n(x,w,b)$ as defined above (as an $N \times 1$ array). 

````julia
function gn(x::Array, w::Vector, b::Number, f_a::Function)
    ## TODO: Fill in ?? to return an Nx1 array with elements given by Equation (2) when x is an n X N array  
    return f_a.(x' * w .+ b)  

end
````

## Formulating binary classification as a learning problem

In the binary classification problem, we are given two classes of variables. Let's call them "Class 0" and "Class 1". In the application we will consider, Class "0" will be the digit 0, and Class "1" will be the digit 9.

The labeling is arbitrary, but once we pick it we must stick with it.

**Setup**:

Now we need training data for the two classes, so our neural network can learn how to distinguish them.

Suppose we have a training dataset comprised of $n$-dimensional vectors $[x_1^{"0"}, x_{2}^{"0"} \ldots ~x_{N_0}^{"0"}]$ asssociated with the Class 0 category. (The superscript in $x_1^{"0"}$ indicates that this is a vector of inputs belonging to Class 0.)

Similarly, we have a training dataset comprised of $n$-dimensional vectors $[x_1^{"1"}, x_{2}^{"1"} \ldots ~x_{N_1}^{"1"}]$ associated with Class 1. (As before, the superscript reminds us that $x_1^{"1"}$ is a vector of inputs belonging to Class 1.)

These two sets of vectors with known classifications contain the "meanings" of Class 0 and Class 1, and we can train a neural network to use these training vectors to distinguish between the two classes.

**Objective**:

Our objective is to train a neural network to distinguish between Class 0 and Class 1. This means that if we are given a vector $x_{\rm test}$ which belongs to Class 0, but is *not* in the original training data set, then the trained neural network will correctly predict that it belongs to Class 0. We cannot guarantee that we will always be able to correctly predict the class for any input, but we hope to train the network to perform this task "reasonably well", which in this case means that it will not errorneously label an input too often. We will quantify this later on.

Having decided on a neural network structure, we are ready to talk about how to train it. The first step in training is to decide on a *loss function*.

## The mean-squared error loss function

One commonly used loss function is **mean-squared-error**:

$$L(x,y,w,b) = \dfrac{1}{N} \sum_{j=1}^{N} \left[y_j - f_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right)\right]^2,$$

Take a moment to make sure you understand the math here. The roles of $x$, $w$ and $b$ are clear from the description of the neural network. What is $y$?

If we remove the $1/N$ normalization in front, we have **squared error**. The parameters that minimize squared error will also minimize mean squared error; however, as you will shortly see, the learning rate needed will depend on $N$ if we use squared error, but not when we used mean squared error.

For the classification problem, $y$ is an **encoding** of the class labels. For example, we might say that Class 0 is associated with the value $-1.0$, and Class 1 is associated with the value +1.0. Other encodings are possible, and we will see later how the encoding affects the decision rule (which takes the output of the neural network and predicts the label of the test vector). For now, just keep in mind what the loss function says: we would like to make $y_j - f_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right)$ small for all $j$ in 1 to $N$.

*Optimal* weights and biases are those that minimize $L$. With optimal choices of $w$ and $b$, we say the network is trained. And a trained network should be able to output a value $f_{\rm a}\left(\sum_{i=1}^{n} w_i x_{i} + b\right)$ which is close to the appropriate encoding value $y$ for an input $x$.

## Gradient of the loss function 

So the next question is: how do we compute optimal $w$ and $b$? The answer comes from calculus. 

To minimize a convex function like $L$, we should try to find values of $w$ and $b$ for which the gradients $\nabla_w L(x,y,w,b)$ and $\nabla_b L(x,y,w,b)$ are zero. When this happens, we have found a "stationary point", and we can declare the corresponding $w$ and $b$ to be optimal.

This means we need to analytically express the *gradient vector* $\nabla_w L(x,y,w,b)$ and *gradient* $\nabla_b L(x,y,w,b)$. Now you see why we need the derivative of the activation function. You've taken derivatives before, so let's skip the somewhat messy derivation and take a look at the results:

$$ \nabla_{w_k} L(x,y,w,b)  = - \dfrac{2}{N}\sum_{j=1}^{N} \left[y_j -  f_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right)\right] f'_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right) \, x_{kj},$$
and

$$ \nabla_{b} L(x,y,w,b)  = - \dfrac{2}{N}\sum_{j=1}^{N} \left[y_j -  f_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right)\right] f'_{\rm a}\left(\sum_{i=1}^{n} w_i x_{ij} + b\right) \, .$$

---

Let's write a function called `loss_gradient` which takes as its inputs `x`, `y`, `w`, `b`, `f_a`, and `df_a` (and optional `normalize` like in the prelecture), and returns three things: the gradient vector $\nabla_{w} L(x,y,w,b)$, the gradient $\nabla_{b} L(x,y,w,b)$, and the loss $L(x,y,w,b)$.

Hint: Vectorize over $k$ to compute all $\nabla_{w_k} L(x,y,w,b)$ together.

````julia
function grad_loss(
        f_a::Function,
        df_a::Function,
        x::Matrix,
        y::Vector,
        w::Vector,
        b::Number,
        normalize::Bool=true
    )
    dw = zeros(length(w))
    db = 0.0
    loss = 0.0
    for j in 1:size(x, 2)
        error =  y[j] - f_a.(w' * x[:, j] + b) ## TODO: fill in ??
        common_term = (error .* df_a.(w' * x[:,j] + b))
        dw = dw .- 2 * common_term .* x[:,j] ## TODO: fill in ?? Ignore the 1/N part that is done below
        db = db .- 2 * common_term  ## TODO: fill in ?? Ignore the 1/N part that is done below
        loss += error^2
     end
    
     if normalize
        dw = dw / length(y)
        db = db / length(y)
        loss = loss / length(y)
    end
    return dw, db, loss
 end
````

## Learning network parameters using gradient descent

Now that we can compute the loss function and its gradient, we are ready to minimize $L$ using gradient descent. If you haven't implemented gradient descent before, get ready to learn a powerful applied mathematics technique! This is an iterative method that starts with an initial value $w_0$ and performs the following update at iteration $k$:

$$w_{k+1} =w_{k} - \mu \,\nabla_w L(x,y,w_k,b_k).$$

Here $\mu>0$ is the *learning rate* (or step size). The subscript $k$ refers to the value of a variable after $k$ steps or iterations. The parameter $b$ is updated similarly via

$$b_{k+1} =b_{k} - \mu \,\nabla_b L(x,y,w_k,b_k).$$

Clearly, we need to know the gradients to be able to perform this update. Luckily you have just written a function to compute them!

Intuitively, when we update $w$ and $b$, we are taking a step "downhill" (of size $\mu$) in an attempt to drive the gradient to zero.

Note: the selection of $\mu$ plays a crucial role: it determines if the networks "learns" or not, and how quickly a stationary point is reached. We will explore this later on.

Complete the function `learn2classify_gd` in the next cell -- the function takes as its inputs `x`,`y`, `f_a`, `df_a`, and `loss_gradient` (and some other keyword parameters), and returns the values of $w$, $b$ and the $L$ after `iters` iterations. 

````julia
using Random: seed!
function learn2classify_gd(
        f_a, 
        df_a, 
        grad_loss, 
        x, 
        y, 
        mu=1e-3, 
        iters=500, 
        show_loss=true, 
        normalize=true, 
        seed=1
    )
    n = size(x, 1)
    
    if seed == false
        w = zeros(n) # vector of zeros
        b = 0.0
    else
        seed!(seed) #initialize random number generator
        w = randn(n)
        b = rand()
    end
    
    loss = zeros(iters) 
    for i in 1:iters
        
        dw, db, loss_i = grad_loss(f_a, df_a, x, y, w, b, normalize)
        w = w - mu * dw 
        b = b - mu * db
        loss[i] = convert(Float64, loss_i[1])
        
        if show_loss
            if(rem(i,100) == 0)
                IJulia.clear_output(true)
                loss_plot = scatter(
                    [1:50:i], loss[1:50:i], yscale=:log10, 
                    xlabel="iteration", 
                    ylabel="training loss", 
                    title="iteration $i, loss = $loss_i"
                ) 
                display(loss_plot)
#                 sleep(0.1)
            end
        end
    end
    return w, b, loss
end
````

## Faster learning with stochastic gradient descent

So far we have evaluated the gradient and loss function over the entire training set. This can be slow if there are many training samples; the training dataset may even be too large to fit in the memory of a single machine. Stochastic Gradient Descent (SGD) addresses this by evaluating the gradient and loss function only for *randomly chosen subsets* (called "mini batches") of the training data at each iteration.

With SGD, we compute each iteration's value of $w$ with

$$w_{k+1} = w_{k} - \mu \,\nabla_w \textrm{L}(\color{blue}{\textrm{ random mini batch of }}{x},\color{blue}{\textrm{ random mini batch of }}{y},w_k,b_k),$$

and an analogous iteration for $b$. At every step, a new random mini batch of the training dataset is selected -- this greatly reduces the computational burden of training the neural network, particularly when there are many training samples.

````julia
using Random: seed!, randperm
function learn2classify_sgd(
        f_a, 
        df_a,
        grad_loss,
        x,
        y,
        mu=1e-3,
        iters=10,
        batch_size=10,
        show_loss=true,
        normalize=true,
        seed=1
    )
    
    n = size(x, 1)
    N = size(x, 2) 
    
     if seed == false
        seed!(1)
        w = zeros(n)
        b = 0.0
    else
        seed!(seed)
        w = randn(n)
        b = rand()
    end

    loss = zeros(iters) 
    for i in 1:(iters)
        batch_idx = randperm(N)
        batch_idx = batch_idx[1:min(batch_size, N)]
        dw, db, loss_i = grad_loss(f_a, df_a, x[:, batch_idx], y[batch_idx], w, b, normalize)
        w = w - mu * dw
        b = b - mu * db
        loss[i] = convert(Float64, loss_i[1])
        
        if show_loss == true
            if(rem(i, 100) == 0)
                IJulia.clear_output(true)
                loss_plot = scatter(
                    [1:50:i], loss[1:50:i], yscale=:log10, 
                    xlabel="iteration", 
                    ylabel="training loss", 
                    title="iteration $i, loss = $loss_i"
                ) 
                display(loss_plot)
#                 sleep(0.1)
            end
        end
    
    end
    return w, b, loss
end
````

## Nesterov's accelerated gradient descent method

Nesterov invented an ingenious method for accelerating gradient descent by modifying the iteration to include a so-called *momentum* term. This involves computing

$$
q_{k+1} = w_{k} - \mu \,\nabla_w L(\textrm{x},\textrm{y},w_k,b_k),
$$

and then computing the $w$ update via the iteration 

\begin{equation} \label{eq:ngd 2}
w_{k+1} = (1-\gamma_k)\, q_{k+1} + \gamma_k\, q_k.
\end{equation}

The scalar parameter $\gamma_k$ is defined as

$$\gamma_k = \dfrac{1-\lambda_k}{\lambda_{k+1}},$$

where

$$\lambda_0 = 0 \textrm{ and } \lambda_{k+1} = \dfrac{1+\sqrt{1+4\,\lambda_{k}^2}}{2}.$$

A similar scheme is used to update $b$. The same idea can be applied to accelerate the stochastic gradient descent method by updating over mini-batches instead. 

___

Implement the accelerated stochastic gradient algorithm (ASGD) in the function `learn2classify_asgd` by filling in the `??` in the next cell.

````julia
using Random: seed!, randperm
function learn2classify_asgd(
        f_a,
        df_a,
        grad_loss,
        x,
        y,
        mu=1e-3,
        iters=500,
        batch_size=10,
        show_loss=true,
        normalize=true,
        seed=1
    )
    
    n = size(x, 1)
    N = size(x, 2)

    if seed == false
        b = 0.0
        w = zeros(n)
    else
        seed!(seed) # initiliaze random number generator
        w = randn(n)
        b = rand()
    end

    loss = zeros(iters)
    
    lambdak = 0
    qk = w
    pk = b
    for i in 1:iters
        batch_idx = randperm(N)
        batch_idx = batch_idx[1:min(batch_size, N)]

        dw, db, loss_i = grad_loss(f_a, df_a, x[:, batch_idx], y[batch_idx], w, b, normalize)
        qkp1 = w - mu * dw
        pkp1 = b - mu * db

        lambdakp1 = (1 + sqrt(1 + 4 * lambdak^2)) / 2
        gammak = (1 - lambdak) / lambdakp1

        w = (1 - gammak) * qkp1 + gammak * qk
        b = (1 - gammak) * pkp1 + gammak * pk

        qk = qkp1
        pk = pkp1
        lambdak = lambdakp1

        loss[i] = convert(Float64, loss_i[1])

        if show_loss == true
            if(rem(i,100) == 0)
                IJulia.clear_output(true)
                loss_plot = scatter(
                    [1:50:i], loss[1:50:i], yscale=:log10,
                    xlabel="iteration",
                    ylabel="training loss",
                    title="iteration $i, loss = $loss_i"
                )
                display(loss_plot)
#                 sleep(0.1)
            end
        end
    end
    return w,b, loss
end
````

We now examine the learning rate using the accelerated gradient descent method.

````julia
@time w_hat, b_hat, loss = learn2classify_asgd(linear, dlinear, grad_loss, x, y, mu, 500, 10, true, true, false);
````

Theory says that in the worst-case, ASGD will converge faster than SGD. It does not say that ASGD will always converge faster. We will shortly see that for a real-world application, ASGD will converge much faster than SGD.

````julia
function generate_test_train_set_from_datacube(
        data::Array,
        percentage_train::Number=50.0
    )
    
    num_samples = size(data, 3)
    num_train = Integer(round(percentage_train / 100.0 * num_samples))
    num_test = num_samples - num_train
    
    rand_idx = randperm(num_samples)
    train_idx = rand_idx[1:num_train]
    test_idx = rand_idx[(num_train + 1):end]
    
    train_data = data[:, :, train_idx]
    test_data = data[:, :, test_idx]
    
    return train_data, test_data
end
````

## Decision Rule and how the class encoding changes the answer

**How do we make a prediction using this output?**

Recall that we encoded Class 0 and Class 1 as the first and second elements of the `class_encoding_vector`. Let 

$$ \textrm{class_encoding_midpoint} = \dfrac{\textrm{class_encoding_vector[1] + class_encoding_vector[2]}}{2},$$

denote the midpoint of the two class encodings.

Then the simplest classifier predicts the class label for the test vector by determining if $g_{\rm a}(x_{\rm test}, \widehat{w}, \widehat{b})$ is closer to $\textrm{class_encoding[1]}$ or  $\textrm{class_encoding[1]}$. If the former, then we declare the test vector to belong to Class 0 and as Class 1 if the latter. If equally close, we choose a class identity by flipping a coin. We will discuss more complicated decision rules and the statistical bases for them in subsequent chapters (Hint: Bayes' rule plays a vital role).

This decision rule can be succinctly encoded mathematically as 

\begin{equation}
g_{\rm a}(x_{\rm test},\widehat{w},\widehat{b}) -\textrm{class_encoding_midpoint} < 0  \Rightarrow \textrm{Classify vector as belong to "Class 0"}
\end{equation}

\begin{equation}
g_{\rm a}(x_{\rm test},\widehat{w},\widehat{b}) -\textrm{class_encoding_midpoint} > 0  \Rightarrow \textrm{Classify vector as belong to "Class 1"} 
\end{equation}

When we have encoded the labels into -1 and +1, as we have done thus far, then the decision rule is particularly simple and given by

\begin{equation}
\textrm{predicted encoding of test vector} = \textrm{sign} \left[g_{\rm a}(x_{\rm test},\widehat{w},\widehat{b})\right]
\end{equation}

This decision rule is encoded in the next cell and we can compute the label predictions for the entire test data set using the `sign` function and the `.` operator. How well does the network predict digits?

````julia
function compute_pcorrect(
        digits::Vector,
        f_a::Function,
        df_a::Function,
        class_encoding_vector::Vector=[-1, 1]
        )
    
    x0, x1 = load_digit_data(digits)
    
    ### TODO: complete based on code from above
    x0_train, x0_test = generate_test_train_set_from_datacube(x0)
    x1_train, x1_test = generate_test_train_set_from_datacube(x1)
    
    x0_train_matrix = datacube2matrix(x0_train) 
    x1_train_matrix = datacube2matrix(x1_train)
    x0_test_matrix = datacube2matrix(x0_test)
    x1_test_matrix = datacube2matrix(x1_test)
    
    train_matrix, train_vector = encode_class0_class1_data(x0_train_matrix, x1_train_matrix, class_encoding_vector)
    test_matrix, test_vector = encode_class0_class1_data(x0_test_matrix, x1_test_matrix, class_encoding_vector)
    
    show_loss = false
    mu = 1e-11 # use the same one from earlier 
    batch_size = 20
    iters = 5000
    w_hat, b_hat, loss = learn2classify_asgd(
        f_a, df_a, grad_loss, 
        train_matrix, train_vector, mu,
        iters, batch_size, show_loss, true, false
        );
    test_prediction = sign.(gn(test_matrix, w_hat,b_hat, f_a)) 
    return pcorrect = sum(test_prediction .== test_vector) / length(test_vector) * 100
end
````

We will utilize the so-called "ternary operator" in Julia to make the code simple, legible and just plain beautiful.

The ternary operator is closely related to the `if-elseif-else` syntax, but is used where a conditional choice between single expression values is required, as opposed to conditional execution of longer blocks of code. It gets its name from being the only operator in most languages taking three operands.

It is invoked as 
```julia
a ? b : c
```

The expression `a`, before the `?`, is a condition expression. The ternary operation evaluates the expression `b` (before the :) if the condition `a` is true, but evaluates the expression `c` (after the `:`) if `a` is false.

Thus 
```julia
(1 < 2) ? println("yes") : println("no")
```

will return the output `yes` because `1` is less than `2`.

The ternary operator will be useful because  classification matrix is symmetric with the diagonals equal to 1 (why?). Thus, only the values in the upper diagonal need to be calculated -- the remaining entries can be populated without running the simulation.

````julia
##TODO Compute and display the classifcation matrix for the modified decision rule with the sigmoid activation
   
function compute_pcorrect1(
        digits::Vector,
        f_a::Function,
        df_a::Function,
        class_encoding_vector::Vector=[-1, 1]
        )
    
    x0, x1 = load_digit_data(digits)
    
    ### TODO: complete based on code from above
    x0_train, x0_test = generate_test_train_set_from_datacube(x0)
    x1_train, x1_test = generate_test_train_set_from_datacube(x1)
    
    x0_train_matrix = datacube2matrix(x0_train) 
    x1_train_matrix = datacube2matrix(x1_train)
    x0_test_matrix = datacube2matrix(x0_test)
    x1_test_matrix = datacube2matrix(x1_test)
    
    train_matrix, train_vector = encode_class0_class1_data(x0_train_matrix, x1_train_matrix, class_encoding_vector)
    test_matrix, test_vector = encode_class0_class1_data(x0_test_matrix, x1_test_matrix, class_encoding_vector)
    
    show_loss = false
    mu = 1e-7 # use the same one from earlier 
    batch_size = 20
    iters = 5000
    w_hat, b_hat, loss = learn2classify_asgd(
        f_a, df_a, grad_loss, 
        train_matrix, train_vector, mu,
        iters, batch_size, show_loss, true, false
        );
        test_prediction = sign.(gn(test_matrix, w_hat,b_hat, f_a).-0.5) 
    for i in 1:length(test_prediction)
        if test_prediction[i] == -1
        test_prediction[i] = 0
        end
    end
    return pcorrect = sum(test_prediction .== test_vector) / length(test_vector) * 100
end

f_a = sigmoid
df_a = dsigmoid
msigmoid_classify_matrix_upper_diagonal = 
    [(digit1 < digit2) ? compute_pcorrect1([digit1, digit2], f_a, df_a, [0, 1]) : 0.0 for digit1 in 0:9, digit2 in 0:9]
msigmoid_classify_matrix = msigmoid_classify_matrix_upper_diagonal + msigmoid_classify_matrix_upper_diagonal' + 100 * I
@show heatmap(msigmoid_classify_matrix; yflip=true)
@show msigmoid_classify_matrix
````
